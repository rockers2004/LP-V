```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('bfs.csv')

# Create simple plots


# Plot 1: Execution Time
plt.figure()
plt.plot(df['size'], df['seq_time'], marker='o', linewidth=2, markersize=8, label='Sequential')
plt.plot(df['size'], df['par_time'], marker='s', linewidth=2, markersize=8, label='Parallel')
plt.xlabel('Graph Size')
plt.ylabel('Time (seconds)')
plt.title('Execution Time')
plt.legend()
plt.grid(alpha=0.3)

# Plot 2: Speedup
plt.figure
plt.plot(df['size'], df['speedup'], marker='o', linewidth=2, markersize=8)
plt.xlabel('Graph Size')
plt.ylabel('Speedup')
plt.title('Speedup')
plt.grid(alpha=0.3)

# Plot 3: Efficiency
plt.figure()
plt.plot(df['size'], df['efficiency'], marker='s', linewidth=2, markersize=8, color='red')
plt.xlabel('Graph Size')
plt.ylabel('Efficiency')
plt.title('Efficiency')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('bfs_performance.png', dpi=150)
plt.show()

print("Plot saved!")
```
